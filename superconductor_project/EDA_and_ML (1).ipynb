```python
!pip install -q pandas numpy matplotlib seaborn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set_style("whitegrid")
sns.set_context("notebook")

FILE = "train.csv.zip"

df = pd.read_csv(FILE)
print("Rows, Columns:", df.shape)
display(df.head())


print("\nINFO:")
print(df.info())

print("\nMissing values per column:")
print(df.isna().sum().sort_values(ascending=False).head(10))

print("\nDuplicates:", df.duplicated().sum())


print("\nDescriptive Statistics:")
display(df.describe().T)


TARGET = "critical_temp"

plt.figure(figsize=(7,4))
sns.histplot(df[TARGET], kde=True, bins=40)
plt.title("Distribution of Critical Temperature")
plt.tight_layout()
plt.show()

print("Skew:", df[TARGET].skew())
print("Kurtosis:", df[TARGET].kurtosis())


corr = df.corr()
corr_target = corr[TARGET].abs().sort_values(ascending=False)
top15 = corr_target.index[:15]

plt.figure(figsize=(10,8))
sns.heatmap(df[top15].corr(), annot=True, cmap="vlag", fmt=".2f", center=0)
plt.title("Top 15 Feature Correlations with Critical Temperature")
plt.tight_layout()
plt.show()

print("\nTop 15:")
print(corr_target.head(15))


num_cols = df.select_dtypes(include=[np.number]).columns.drop(TARGET)
top_var = df[num_cols].var().sort_values(ascending=False).index[:12]

plt.figure(figsize=(14,10))
for i, col in enumerate(top_var, 1):
    plt.subplot(3,4,i)
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()


outlier_info = {}
for col in num_cols:
    q1, q3 = df[col].quantile([0.25, 0.75])
    iqr = q3 - q1
    low = q1 - 1.5*iqr
    high = q3 + 1.5*iqr
    outlier_info[col] = ((df[col] < low) | (df[col] > high)).sum()

out_series = pd.Series(outlier_info).sort_values(ascending=False)
print("\nTop outlier-heavy features:")
display(out_series.head(10))

print("\nEDA COMPLETE")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd



TARGET = "critical_temp"


df["target"] = np.log1p(df[TARGET])


df["target_bin"] = pd.qcut(df["target"], q=10, duplicates="drop")


X = df.drop(columns=[TARGET, "target", "target_bin"])
y = df["target"]


X_train, X_tmp, y_train, y_tmp = train_test_split(
    X, y, test_size=0.40, random_state=42, stratify=df["target_bin"]
)
X_val, X_test, y_val, y_test = train_test_split(
    X_tmp, y_tmp, test_size=0.50, random_state=42
)


scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_val_s = scaler.transform(X_val)
X_test_s = scaler.transform(X_test)

print("Train:", X_train_s.shape, y_train.shape)
print("Validation:", X_val_s.shape, y_val.shape)
print("Test:", X_test_s.shape, y_test.shape)
```

```python

import gc, sys, time
import numpy as np, pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


for name in list(globals().keys()):
    if name.startswith("rf_search") or name.startswith("gbr_search") or name.startswith("xgb_search"):
        try:
            del globals()[name]
        except Exception:
            pass
gc.collect()
print("Cleanup done. Free mem hint:", gc.get_threshold())


def inv_exp(x): return np.expm1(x)
def eval_model(model, Xv, yv):
    preds = model.predict(Xv)
    y_true = inv_exp(yv)
    y_pred = inv_exp(preds)
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(((y_true - y_pred)**2).mean())
    r2 = r2_score(y_true, y_pred)
    return {"MAE": mae, "RMSE": rmse, "R2": r2}


FRACTION = 0.25
rs = 42
n_train = X_train_s.shape[0]
if FRACTION < 1.0:
    idx = np.random.RandomState(rs).choice(n_train, size=max(200, int(n_train*FRACTION)), replace=False)
    Xs_sub = X_train_s[idx]
    ys_sub = y_train.iloc[idx].reset_index(drop=True)
else:
    Xs_sub = X_train_s
    ys_sub = y_train.reset_index(drop=True)

print("Using subsample rows:", Xs_sub.shape[0])


t0 = time.time()
lr = LinearRegression()
lr.fit(Xs_sub, ys_sub)
lr_res = eval_model(lr, X_val_s, y_val)
t1 = time.time()
lr_res['time_s'] = round(t1-t0,2)
print("LinearRegression ->", lr_res)


t0 = time.time()
rf = RandomForestRegressor(n_estimators=80, max_depth=10, random_state=rs, n_jobs=1)
rf.fit(Xs_sub, ys_sub)
rf_res = eval_model(rf, X_val_s, y_val)
t1 = time.time()
rf_res['time_s'] = round(t1-t0,2)
print("RandomForest (small) ->", rf_res)


t0 = time.time()
xgb = XGBRegressor(n_estimators=100, learning_rate=0.05, max_depth=3,
                   objective="reg:squarederror", tree_method="hist", random_state=rs, n_jobs=1)
xgb.fit(Xs_sub, ys_sub)
xgb_res = eval_model(xgb, X_val_s, y_val)
t1 = time.time()
xgb_res['time_s'] = round(t1-t0,2)
print("XGBoost (small) ->", xgb_res)


results = {"LinearRegression": lr_res, "RandomForest": rf_res, "XGBoost": xgb_res}
best_name = min(results.keys(), key=lambda k: results[k]['MAE'])
print("Best quick model:", best_name)

best_model = {"LinearRegression": lr, "RandomForest": rf, "XGBoost": xgb}[best_name]

print("Refitting best model on full train+val and evaluating on test (this is quick)...")
X_full_train = np.vstack([X_train_s, X_val_s])
y_full_train = np.concatenate([y_train, y_val])
t0 = time.time()
best_model.fit(X_full_train, y_full_train)
t1 = time.time()
test_res = eval_model(best_model, X_test_s, y_test)
test_res['time_s_refit'] = round(t1-t0,2)
print("Test eval (best model):", test_res)


import joblib, os
os.makedirs("outputs/models", exist_ok=True)
joblib.dump(best_model, "outputs/models/best_quick_model.joblib")
print("Saved quick model -> outputs/models/best_quick_model.joblib")
```

```python
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

os.makedirs("outputs/figures", exist_ok=True)


model = joblib.load("outputs/models/best_quick_model.joblib")

feature_names = X_train.columns.tolist()

if hasattr(model, "feature_importances_"):
    imp = model.feature_importances_
else:
    imp = np.abs(model.coef_)

feat_imp = pd.Series(imp, index=feature_names).sort_values(ascending=False)
print("Top 20 features by importance:")
print(feat_imp.head(20))

# plot top 20
plt.figure(figsize=(8,10))
feat_imp.head(20).sort_values().plot.barh()
plt.title("Top 20 feature importances")
plt.tight_layout()
plt.savefig("outputs/figures/feature_importances_top20.png")
plt.show()

```

```python
from sklearn.metrics import mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import numpy as np

# predictions on test
preds_log = model.predict(X_test_s)           # predictions in log1p-space
y_test_orig = np.expm1(y_test)                # true target in original Tc-space
preds_orig = np.expm1(preds_log)              # predicted Tc in original space

# metrics
mae = mean_absolute_error(y_test_orig, preds_orig)
rmse = np.sqrt(((y_test_orig - preds_orig)**2).mean())
r2 = r2_score(y_test_orig, preds_orig)
print(f"Test MAE: {mae:.4f}, RMSE: {rmse:.4f}, R2: {r2:.4f}")

# parity plot
plt.figure(figsize=(6,6))
plt.scatter(y_test_orig, preds_orig, alpha=0.4, s=12)
mn = min(y_test_orig.min(), preds_orig.min())
mx = max(y_test_orig.max(), preds_orig.max())
plt.plot([mn,mx],[mn,mx],"k--")
plt.xlabel("True critical_temp")
plt.ylabel("Predicted critical_temp")
plt.title("Parity plot")
plt.tight_layout()
plt.savefig("outputs/figures/parity_plot.png")
plt.show()

# residuals
residuals = y_test_orig - preds_orig
plt.figure(figsize=(6,4))
plt.scatter(preds_orig, residuals, alpha=0.4, s=12)
plt.hlines(0, xmin=preds_orig.min(), xmax=preds_orig.max(), colors="k", linestyles="dashed")
plt.xlabel("Predicted critical_temp")
plt.ylabel("Residual (true - pred)")
plt.title("Residuals vs Predicted")
plt.tight_layout()
plt.savefig("outputs/figures/residuals_vs_pred.png")
plt.show()
```

```python

from flask import Flask, request, jsonify
import joblib
import pandas as pd
import numpy as np

app = Flask(__name__)
pipeline = joblib.load("outputs/models/final_pipeline.joblib")
scaler = pipeline["scaler"]
model = pipeline["model"]
feature_names = pipeline["feature_names"]

@app.route("/predict", methods=["POST"])
def predict():
    payload = request.get_json(force=True)
    # accept single dict or list of dicts
    single = False
    if isinstance(payload, dict):
        data = pd.DataFrame([payload])
        single = True
    else:
        data = pd.DataFrame(payload)

    data = data.reindex(columns=feature_names, fill_value=0.0)
    X_scaled = scaler.transform(data)
    preds_log = model.predict(X_scaled)
    preds = np.expm1(preds_log).tolist()
    return jsonify({"predictions": preds} if not single else {"prediction": preds[0]})

if __name__ == "__main__":
    app.run(host="0.0.0.0", port=5000)
```
